```python
def x_normal_distrubition_function(x):
    value = x * scipy.stats.norm.pdf(x, loc = 1 ,scale = 2 )
    return value

scipy.integrate.quad(x_normal_distrubition_function, -np.inf , np.inf)
```

```python
def v_normal_distrubition_function(x):
    value =( x-1 )**2 *  scipy.stats.norm.pdf(x, loc = 1 ,scale = 2 )
    return value

scipy.integrate.quad(v_normal_distrubition_function, -np.inf , np.inf)
```
